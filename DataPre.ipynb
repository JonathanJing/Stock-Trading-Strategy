```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/isds577/keystats.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df1 = df[['calendardate','ticker',
    'eps','netinc','opinc','ncfo','cashnequsd',
          'de','ebitda','fcf','marketcap','netmargin',
          'pb','pe','ps','workingcapital','ev',
          'divyield','fcfps','revenue']]
```

```python
df1.head(10)
```

```python
df1.shape
```

```python
#Find zeros in columns
(df1==0).astype(bool).sum(axis=0)
```

```python
#Find NAs in columns
print(df1.isnull().sum())
```

```python
df.earnings = df1[['eps','netinc','opinc','ncfo','ebitda',
                 'fcf','netmargin','divyield','fcfps','revenue']]
df.ratio = df1[['pb','pe','ps']]
df.value = df1[['workingcapital','ev','marketcap','cashnequsd']]
df.other = df1['de']
```

```python
df.earnings = df.earnings.fillna(0)
df.ratio = df.ratio.fillna(0)
df.other = df.other.fillna(0)
df.value = df.value.fillna(df.value.mean())
```

```python
new = pd.concat([df.ratio,df.earnings,df.other,df.value],axis=1)
```

```python
new.head(10)
```

```python
# Recheck NAs
print(new.isnull().sum())
```

```python
#ROE, ROA, ROICCalculation
df2 = df[['equityusd','netinc','assets','invcap','ebit']]
```

```python
df2.head(10)
```

```python
df2 = df2.fillna(0)
```

```python
# Recheck NAs
print(df2.isnull().sum())
```

```python
df2['roe']= df2['netinc'] / df['equityusd']
df2['roa'] = df2['netinc'] / df['assets']
df2['roic'] = df2['netinc'] / df['invcap']
```

```python
df2.head(10)
```

```python
ratio_cal = df2[['roe','roa','roic']]
```

```python
# Recheck NAs
print(ratio_cal.isnull().sum())
ratio_cal=ratio_cal.fillna(0)
```

```python
#ticker, dim, etc.
info = df.iloc[:,0:6]
info.head(10)
```

```python
#Combine info, ratio_cal, and new.
keystats_new = pd.concat([info, new, ratio_cal], axis =1)
```

```python
keystats_new.shape
```

```python
# Recheck NAs
print(keystats_new.isnull().sum())
```
