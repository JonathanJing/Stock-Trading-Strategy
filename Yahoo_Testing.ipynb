```python
def get_financial_data(stock, period):
    balance = YahooFinancials(stock).get_financial_stmts(period, "balance")
    income = YahooFinancials(stock).get_financial_stmts(period, "income")
    cash = YahooFinancials(stock).get_financial_stmts(period, "cash")
    return (balance, income, cash)

def get_dfp(accountability, stock):
    dfps = pd.DataFrame()
    years = []
    for history in accountability:
        for time_avaliable in accountability[history][stock]:
            for year in time_avaliable:
                dataframe = pd.DataFrame(data=time_avaliable[year], columns=time_avaliable[year].keys(), index=[year])
                years.append(year)
                if(dfps.empty):
                    dfps = dataframe.copy()
                else:
                    dfps = dfps.append(dataframe, ignore_index=True)
    dfps.insert(0, "year", years)
    return dfps

def get_dfps(stock, period):
    stock = str(stock).upper() + ".SA"
    (balance, income, cash) = get_financial_data(stock, period)

    balance = get_dfp(balance, stock)
    income = get_dfp(income, stock)
    cash = get_dfp(cash, stock)

    return [balance, income, cash]
```

```python
[balance, income, cash] = get_dfps('AAPL', 'quarterly')
```
